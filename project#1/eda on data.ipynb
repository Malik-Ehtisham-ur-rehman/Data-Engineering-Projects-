```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV file
df = pd.read_csv('exchange.csv')

```

```python
df.head()
```

```python
# Data Cleaning
df['Total assets (2022) (US$ billion)'] = df['Total assets (2022) (US$ billion)'].str.replace('$', '').str.replace(',', '').astype(float)

```

```python
# Summary Statistics
total_assets = df['Total assets (2022) (US$ billion)'].sum()
average_assets = df['Total assets (2022) (US$ billion)'].mean()
median_assets = df['Total assets (2022) (US$ billion)'].median()
min_assets = df['Total assets (2022) (US$ billion)'].min()
max_assets = df['Total assets (2022) (US$ billion)'].max()

print(f"Total Assets: ${total_assets:.2f} billion")
print(f"Average Assets: ${average_assets:.2f} billion")
print(f"Median Assets: ${median_assets:.2f} billion")
print(f"Min Assets: ${min_assets:.2f} billion")
print(f"Max Assets: ${max_assets:.2f} billion")





```

```python

# Select top 20 banks
top_20_df = df.nlargest(20, 'Total assets (2022) (US$ billion)')

# Bar Chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Total assets (2022) (US$ billion)', y='Bank name', data=top_20_df, palette='viridis')
plt.title('Total Assets of Top 20 Banks (2022)')
plt.xlabel('Total assets (US$ billion)')
plt.ylabel('Bank name')
plt.show()
```

```python
# Select top 10 banks
top_10_df = df.nlargest(10, 'Total assets (2022) (US$ billion)')

# Pie Chart
plt.figure(figsize=(10, 10))
plt.pie(top_10_df['Total assets (2022) (US$ billion)'], labels=top_10_df['Bank name'], autopct='%1.1f%%', startangle=140)
plt.title('Market Share of Total Assets among Top 10 Banks (2022)')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(df['Total assets (2022) (US$ billion)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Total Assets among 100 Banks (2022)')
plt.xlabel('Total assets (US$ billion)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Total assets (2022) (US$ billion)'])
plt.title('Box Plot of Total Assets among 100 Banks (2022)')
plt.xlabel('Total assets (US$ billion)')
plt.show()
```

```python
# Cumulative Distribution Plot
df_sorted = df.sort_values(by='Total assets (2022) (US$ billion)', ascending=False)
df_sorted['Cumulative Assets'] = df_sorted['Total assets (2022) (US$ billion)'].cumsum()

plt.figure(figsize=(12, 6))
sns.lineplot(x=df_sorted['Rank'], y=df_sorted['Cumulative Assets'], marker='o')
plt.title('Cumulative Distribution of Total Assets among Banks (2022)')
plt.xlabel('Bank Rank')
plt.ylabel('Cumulative Total assets (US$ billion)')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Rank', y='Total assets (2022) (US$ billion)', data=df, hue='Total assets (2022) (US$ billion)', palette='viridis', size='Total assets (2022) (US$ billion)', sizes=(20, 200))
plt.title('Scatter Plot of Bank Rank vs Total Assets (2022)')
plt.xlabel('Rank')
plt.ylabel('Total assets (US$ billion)')
plt.legend(title='Total assets (US$ billion)', loc='upper right')
plt.show()
```
